```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras_tuner.tuners import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
from sklearn import linear_model, model_selection, preprocessing, metrics

import os
import warnings
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow as tf


warnings.filterwarnings('ignore')
```

```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test.csv')
```

```python
print(f'Train Data: {data_train.shape}\nTest Data: {data_test.shape}')

```

```python
print("Train Data Sütun Sayısı: ", data_train.shape[0])
print("Train Data Satır Sayısı: ", data_train.shape[1])
print("Test Data Sütun Sayısı: ", data_test.shape[0])
print("Test Data Satır Sayısı: ", data_test.shape[1])
```

```python
#DataFrame'de tüm Satırların gözükmesi için kullandım
pd.set_option("display.max_columns", None)

#Eksik veri kontrolü yapıyorum görüldüğü üzere null değer yok
df = df = pd.read_csv("test.csv")
print("\nEksik Veri Sayısı:\n", df.isnull().sum())

```

```python
# veri setinde her bir satır için tekrarlanan verilerin olup olmadığını kontrol ediyorum.
data_train.duplicated().any()

```

```python
# Aynı işlemi sütunları satır gibi okumak için Transpoze alıp kontrol ediyorum.
duplicated_columns = data_train.columns[data_train.T.duplicated()].tolist()
len(duplicated_columns)
```

```python
# axis = 0 satır axis = 1 sütunları ifade ediyor tekrar eden sütunları çıkarıp yazdırıyorum
data_train = data_train.drop(duplicated_columns, axis = 1)
data_train.shape
```

```python
# Tekrarlayan sütunları test setinden de çıkarıyorum
data_test = data_test.drop(duplicated_columns, axis=1)
data_test.shape
```

```python
#veri setinde activity'de bulunan kategorilerin kaç defa tekrar ettiğini gösteren grafiği çizdiriyorum

plt.figure(figsize=(12,6))
axis=sns.countplot(x="Activity",data=data_train, hue="Activity")
plt.xticks(x=data_train['Activity'])
plt.show()
```

```python
activites = data_train["Activity"].unique()

class_map = {i : val for i, val in enumerate(activites)}
class_map_reverse = {key : val for val, key in class_map.items()}

print(class_map)
```

```python
data_train["Activity"] = data_train["Activity"].apply(lambda x : class_map_reverse[x])
data_test["Activity"] = data_test["Activity"].apply(lambda x : class_map_reverse[x])
```

```python
data_train.drop("subject", axis = 1, inplace = True)
data_test.drop("subject", axis = 1, inplace = True)
```

```python
time_steps = 10

def create_sequences(X, y, time_steps = 5):
    X_, y_ = [], []
    n = X.shape[0]
    for i in np.arange(n - time_steps):
        X_.append(X[i:(i + time_steps)])
        y_.append(y[i + time_steps])
    return np.array(X_), np.array(y_)

X_train_ = data_train.drop("Activity", axis = 1).values
X_test_ = data_test.drop("Activity", axis = 1).values
y_train_ = data_train["Activity"].values
y_test_ = data_test["Activity"].values

X_train, y_train = create_sequences(X_train_, y_train_, time_steps)
X_test, y_test = create_sequences(X_test_, y_test_, time_steps)

print(X_train.shape, X_test.shape)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Input(shape = X_train.shape[1:]),
    tf.keras.layers.LSTM(units = 256, return_sequences = True),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.LSTM(units = 128, return_sequences = True),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.LSTM(units = 64),
    tf.keras.layers.Dense(units = 32, activation = "relu"),
    tf.keras.layers.Dense(units = len(class_map), activation = "softmax")
])

model.compile(optimizer = tf.keras.optimizers.Adam(0.005), loss = tf.keras.losses.SparseCategoricalCrossentropy(), metrics = ["accuracy"])
model.summary()
```

```python
cb = tf.keras.callbacks.EarlyStopping(patience = 10, restore_best_weights = True)
history = model.fit(X_train, y_train, epochs = 100, validation_split = 0.3, callbacks = cb)
```

```python
history_df = pd.DataFrame(history.history)
history_df.plot()
plt.show()

history_df[["loss", "val_loss"]].plot()
plt.xlabel("#Epochs")
plt.ylabel("Cross-Entropy Loss")
plt.show()

history_df[["accuracy", "val_accuracy"]].plot()
plt.xlabel("#Epochs")
plt.ylabel("Accuracy")
plt.show()
```

```python
model.evaluate(X_test, y_test)
```

```python
y_test_pred_probs = model.predict(X_test)
y_test_pred = np.array([np.argmax(probs) for probs in y_test_pred_probs])
```

```python
classifier_report = metrics.classification_report(y_test, y_test_pred, digits = 4)
print(classifier_report)
```

```python
lstm_predictions = model.predict(X_test) 

predicted_classes = lstm_predictions.argmax(axis=1)  

X_test_flat = X_test.reshape(X_test.shape[0], -1)

results_df = pd.DataFrame(X_test_flat) 
results_df['Predicted'] = predicted_classes 
results_df['Actual'] = y_test.flatten()  

results_df.to_csv('lstm_predictions.csv', index=False)
print("'lstm_predictions.csv' kaydedildi.")
```
